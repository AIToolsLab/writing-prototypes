# Migration Plan: Python Prototypes to TypeScript

This notebook outlines the plan to convert the Python-based `writing-prototypes` (Streamlit frontend + FastAPI backend) to a TypeScript full-stack application (React frontend + Node.js backend).

## 1. Project Structure and Findings

- **Frontend**: Currently Streamlit (`app.py`, `simplified.py`).
- **Backend**: Currently FastAPI (`custom_llm.py`) using `transformers` and `torch` for local inference (Gemma-2-9b-it).
- **Target Stack**: TypeScript.
    - **Frontend**: React (likely Vite or Next.js) to match `writing-tools` stack.
    - **Backend**: Node.js (Express) with `onnxruntime-node` OR a Python bridge. *Note: Running large LLMs like Gemma-9b in Node.js via ONNX can be challenging. A hybrid approach (TS Frontend + Python Inference Backend) is recommended for performance, but we will explore TypeScript options.*

## 2. Proposed Architecture

### Frontend (TypeScript + React)
Create a new Vite project in `writing-prototypes/frontend`.
- Reimplement Streamlit text areas and buttons as React components.
- Use `axios` or `fetch` to call the backend API.

### Backend (Option A: Hybrid - Recommended)
- Keep `custom_llm.py` (FastAPI) for inference.
- Create a lightweight Node.js/Express server (if needed for auth/logic) that proxies to FastAPI, OR have the Frontend call FastAPI directly.

### Backend (Option B: Pure TypeScript)
- Use `onnxruntime-node`.
- Requires converting `gemma-2-9b-it` to ONNX format (resource intensive).
- Likely slower than PyTorch.

**Decision**: We will proceed with **Option A** (Hybrid) where the "Application Logic" is moved to TypeScript (Frontend), but the "Inference Engine" remains Python. This fulfills the "Change python code... to typescript code" for the application layer while preserving the "local PyTorch inference" capability which is best suited for Python. *However, if the user insists on a TS backend, we can set up an Express server that spawns a Python process or uses a bridge.*

## 3. Plan

1.  **Initialize Frontend**: `npm create vite@latest frontend -- --template react-ts` inside `writing-prototypes`.
2.  **Install Dependencies**: `npm install` (react, axios, tailwindcss, etc.).
3.  **Port Logic**:
    - Convert `app.py`'s `landing`, `rewrite_with_predictions` to React components.
    - Convert `simplified.py`'s chat interface to a React Chat component.
4.  **Backend Integration**:
    - Ensure `custom_llm.py` allows CORS from the new frontend.
    - (Optional) Create a `backend` folder with Express if we need to server-side render or handle intermediate logic.
